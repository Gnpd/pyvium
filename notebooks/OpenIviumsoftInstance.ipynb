# Open/Close IviumSoft from Python

```python
import subprocess
import time
from pyvium import Core, Pyvium
```

```python
def open_iviumsoft_instance():
    '''opens a new IviumSoft Instance'''
    return subprocess.Popen(
        [r"C:\IviumStat\IviumSoft.exe"],
        stdin=subprocess.DEVNULL,
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
        close_fds=True
    )
```

### We open an IviumSoft instance and save the instance information in a variabe


```python
iviumsoft_instance1 = open_iviumsoft_instance()
time.sleep(1)
iviumsoft_instance2 = open_iviumsoft_instance()
print("------------------------------------------------------------")
print("Created iviumsoft instance(1) with PID {}".format(iviumsoft_instance1.pid))
print("Created iviumsoft instance(2) with PID {}".format(iviumsoft_instance2.pid))
print("------------------------------------------------------------")
```

```python
Pyvium.open_driver()
print("------------------------------------------------------------")
print("active instances: ",Pyvium.get_active_iviumsoft_instances())
print("------------------------------------------------------------")
```

```python
Pyvium.select_iviumsoft_instance(2)
```

```python
Pyvium.connect_device()
```

### Closing the instance 2 (without inmediately killing the process)
#### This is equivalent to closing the iviumsoft window and allows iviumsoft to update the dll info

```python
cmd = "(Get-Process -Id "+str(iviumsoft_instance2.pid)+").CloseMainWindow()"
print(cmd)
```

```python
print(subprocess.run(["powershell", "-Command", cmd], capture_output=True))
time.sleep(5)
```
