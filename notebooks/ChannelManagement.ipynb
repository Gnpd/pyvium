## Workaround for channel selection

```python
from pyvium import Pyvium, Core
import time
```

-1=no IviumSoft; 
0=not connected;
1=available_idle;
2=available_busy;
3=no device available



```python
Core.IV_open()
```

```python
def connect_channel(ch):
    '''Connects the first available device to channel "ch",
    if the channel has not device'''
    Core.IV_SelectChannel(ch)
    return Core.IV_connect(1)[0]
```

```python
connect_channel(7)
```

```python
def disconnect_channel(ch):
    '''Disconnects channel (if the device is not busy)'''
    Core.IV_SelectChannel(ch)
    if Core.IV_getdevicestatus() in [1]:
        result_code= Core.IV_connect(0)[0] 
```

```python
disconnect_channel(2)
```

```python
           
def get_channels_statuses(number_of_channels):
    channel_statuses= []
    for ch in range(1,number_of_channels+1):
        Core.IV_SelectChannel(ch)
        status_code = Core.IV_getdevicestatus()
        sn = ''
        if status_code != 3:
            sn = Core.IV_readSN()[1]
        channel_statuses.append([ch,sn,status_code])
    return channel_statuses
        
    
    
```

```python
def connect_devices_to_channels(device_num,number_of_channels):
    # here we reset the devices list
    get_channels_statuses(number_of_channels)
    active_channels = []
    for ch in range(1,device_num+1):
        connect_channel(ch)
        active_channels.append([ch,Core.IV_readSN()[1],Core.IV_getdevicestatus()])
    return active_channels
        
    
print(connect_devices_to_channels(3,4))    

        
```

```python
def connect_device_to_channel(serialNum, channel):
    #check the actual list
    ch_statuses = get_channels_statuses(3)
    print('initial statuses',ch_statuses)
    #if the sn is in the actual list rise an exception: sn already busy
    
    #connect to the channel
    new_ch_statuses = connect_devices_to_channels(channel,3)
    print('intermediate statuses',ch_statuses)
    
    #disconnect the unused new conections
    for index, ch in enumerate(new_ch_statuses):
        if ch[1] != serialNum and ch[1] != ch_statuses[index][1]:
           disconnect_channel(ch[0])
    
    Core.IV_SelectChannel(channel)
    ch_statuses = get_channels_statuses(3)
    
    return ch_statuses
    
    #connect_devices_to_channel(channel)

connect_device_to_channel('R31592', 2)
```
